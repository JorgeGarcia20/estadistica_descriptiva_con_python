# Medidas de dispersión
Las medidas de dispersión tratan, a través del cálculo de diferentes fórmulas, de arrojar un valor numérico que ofrezca información sobre el grado de variabilidad de una variable.

En otras palabras, las medidas de dispersión son números que indican si una variable se mueve mucho, poco, más o menos que otra. 

La razon de ser de este tipo de medidas es conocer de manera resumida una característica de la variable sustituida. En este sentido, deben acompañar a las medidas de tendencia central. Juntas, ofrecen información de un solo vistazo que luego se pueden utilizar para comparar y, si fuera preciso, tomar decisiones.

# Principales medidas de dispersión
Las medidas de dispersión más conocidas son:
## Rango
Es un valor numerico que indica la diferencia entre el valor maximo y mínimo de una población o muestra estadistica. Su fórmula es:
$$R = Max_{x} - Min_{x}$$

Donde:
* $R \to$ es el rango.
* $Max \to$ es el valor máximo de la muestra o población.
* $Min \to$ es el valor mínimo de la muestra o población.
* $x \to$ es la variable sobre la que se pretende calcular esta medida.

## Varianza
La varianza es una medida de dispersion que representa la variabilidad de una serie de datos respecto a su media. Formalmente se calcula como la suma de los residuos al cuadrado divididos entre el total de observaciones. 
Su formula es la siguiente:

$$
\sigma²=\frac{\sum^N_{1}(x_{i}-X̅)²}{N}
$$

$$
\sigma²=\frac{1}{n}\sum^n_{i=1}(x_{i}-\mu)²
$$

donde:
* $ X \to $ es la variable sobre la que se pretenden calcular la varianza.
* $ x_{i} \to $ es la observación número $i$ de la variable $X$. $i$ tomara valores entre $1$ y $n$.
* $ N \to $ es el número de observaciones.
* $ X̅ \to $ es la media de la variable $X$.

## Desviación tipica o desviación estandar
La desviación tipoa es otra medida que ofrece información de la dispercion respecto a la media. Su calculo es exactamente el mismo que la varianza, pero realizando la raíz cuadrada de su resultado. Es decir, la desviación típida es la raíz cuadrada de la varianza.

$$
\sigma=\sqrt{\frac{\sum^N_{1}(x_{i}-X̅)²}{N}}
$$

$$
\sigma=\sqrt{\frac{1}{n-1}\sum^n_{i=1}(x_{i}-\mu)²}
$$

donde:
* $ X \to $ es la variable sobre la que se pretenden calcular la varianza.
* $ x_{i} \to $ es la observacion numero $i$ de la variable $X$. $i$ tomara valores entre $1$ y $n$.
* $ N \to $ es el número de observaciones.
* $ X̅ \to $ es la media de la variable $X$.


Es la medida de dispersion de datos más utilizada. Cuando se esta trabajando con muestras en vez de la problación, se hace una correccion haciendo que se divida para $n-1$.


## Coeficiente de variación
Su calculo se obtiene de dividir la desviación típica entre el valor absoluto de la media del conjunto y por lo general se expresa en porcentaje para su mejor comprensión.

$$CV=\frac{\sigma_{x}}{|X̅|}$$

donde:
* $X \to$ es la variable sobre la que se pretende calcular la varianza.
* $\sigma_{x} \to$ es la desviación típica de la variable $X$.
* $|X̅| \to$ es la media de la variable X en valor absoluto con $ X̅ \neq 0$

El diagrama de caja es la visualización para presentar simplificadamente la dispersion de los datos en referencia a la mediana.

## Medidas de dispersión en Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/cars.csv')
df.shape
```

**Calculando la desviación estándar**

```python
df['price_usd'].std()
```

$\sigma=6428.1520$

**Calculando el rango** = valor máximo - valor mínimo

```python
rng = df['price_usd'].max()-df['price_usd'].min()
rng
```

**Calculando los Quartiles**

```python
prices = df['price_usd']
median = prices.median()
Q1 = prices.quantile(q=0.25)
Q3 = prices.quantile(q=0.75)
min_val = prices.quantile(q=0)
max_val = prices.quantile(q=1)

print(min_val, Q1, median, Q3, max_val)
```

```python
iqr = Q3-Q1
iqr
```

## Límites para detección de outliers (con datos simétricamente distribuídos)

```python
minlimit = Q1 - 1.5*iqr
maxlimit = Q3 - 1.3*iqr
print(minlimit, maxlimit)
```

El valor es negativo porque se está aplicando una ecuacion de una distribución simétrica a una no simétrica.
